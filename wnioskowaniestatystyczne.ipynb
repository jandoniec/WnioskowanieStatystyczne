```python
import numpy as np

x = np.array([-1, 0, 1, 3, 6])

p = np.array([0.1, 0.3, 0.1, 0.2, None])

c = 1 - np.sum(np.array([0.1, 0.3, 0.1, 0.2]))
p[-1] = c

print("c =", c)
print("Zbiór wartości X:", x.tolist())
print("Czy X jest dyskretna? -> tak, ma skończony zbiór wartości")
```

```python
# a) P(X ∈ {−1, 6})
pa = p[(x == -1) | (x == 6)].sum()

# b) P(0 < X ≤ 3)
pb = p[(x > 0) & (x <= 3)].sum()

# c) P(X ≠ 0)
pc = p[x != 0].sum()

# d) P(X = 2)
pd = p[x == 2].sum()    # wynik: 0, bo 2 nie ma w tabeli

# e) P(X < 2)
pe = p[x < 2].sum()

print("a) P(X ∈ {-1, 6}) =", pa)
print("b) P(0 < X ≤ 3) =", pb)
print("c) P(X ≠ 0) =", pc)
print("d) P(X = 2) =", pd)
print("e) P(X < 2) =", pe)
```

```python
F = np.cumsum(p)

for val, fv in zip(x, F):
    print(f"F(x) dla x ≥ {val} wynosi {fv}")

```

```python
# wartość oczekiwana
E = np.sum(x * p)

# wariancja
Var = np.sum((x - E)**2 * p)

# odchylenie standardowe
SD = np.sqrt(Var)

print("Wartość oczekiwana E(X):", E)
print("Wariancja Var(X):", Var)
print("Odchylenie SD(X):", SD)
```
