```python
import numpy as np

# Punkty skoków dystrybuanty
x = np.array([-5, -1, 2, 5])

# Dystrybuanta w tych punktach
F = np.array([0.1, 0.4, 0.6, 1.0])

# Wartość F tuż przed skokiem (poprzedni poziom)
F_prev = np.array([0.0, 0.1, 0.4, 0.6])

# Obliczam prawdopodobieństwa p(x)
p = F - F_prev

print("Zbiór wartości X:", x.tolist())
print("Prawdopodobieństwa p(x):", p.tolist())

for xi, pi in zip(x, p):
    print(f"P(X = {xi}) = {pi}")
```
